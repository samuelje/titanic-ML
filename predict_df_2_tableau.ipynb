```python
#dependencies
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read in train csv
titanic_df = pd.read_csv("predict_df.csv")
conditions = [titanic_df['Prediction'] == titanic_df['Actual'], titanic_df['Prediction'] != titanic_df['Actual']]
choices = ['Correct', 'Incorrect']
titanic_df['Prediction'] = titanic_df['Prediction'].astype(str)
titanic_df['Actual'] = titanic_df['Actual'].astype(str)
titanic_df['Prediction'] = titanic_df['Prediction'].replace(['0','1'],['Perished','Survived'])
titanic_df['Actual'] = titanic_df['Actual'].replace(['0','1'],['Perished','Survived'])
titanic_df['Accuracy'] = np.select(conditions, choices)
#titanic_df.to_csv('tableau_formatted_prediction.csv', index = False)
titanic_df
```
